# Importing data

```python
import pandas as pd
dataset = pd.read_csv("L.Victoria water quality paramenters.csv")
dataset
```

```python
min_value_chlorophyll = dataset['Chlorophyll_content'].min()
max_value_chlorophyll = dataset['Chlorophyll_content'].max()

print(f"Minimum value of chlorophyll: {min_value_chlorophyll} Maximim Value chlorophyll: {max_value_chlorophyll}")
```

# Preparing the data 

## Defining Chlorophyll(y variable) and water salinity and pH(x variables)

```python
y = dataset['Chlorophyll_content']
y
```

```python
X = dataset[['Water_salinity', 'Water_pH']]
X
```

## Splitting between test data and train data

```python
    from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=100)
```

```python
X_train
```

```python
X_test
```

# Model Buiding

## Training the model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

## Prediction using model

```python
y_lr_pred_train = lr.predict(X_train)
y_lr_pred_test = lr.predict(X_test)
```

## Evaluating model performace - comparing the y_lr_pred_train that has been predicted with what was originally there in dataset

```python
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, root_mean_squared_error
```

```python
y_lr_train_mse = mean_squared_error(y_train,y_lr_pred_train)
y_lr_train_r2 = r2_score(y_train, y_lr_pred_train)
y_lr_train_rmse = root_mean_squared_error(y_train, y_lr_pred_train)

print("Mse on train data :", y_lr_train_mse)
print("R2 score on train data: ", y_lr_train_r2)
print("Rmse score on train data: ", y_lr_train_rmse)
```

```python
y_lr_test_mse = mean_squared_error(y_test,y_lr_pred_test)
y_lr_test_r2 = r2_score(y_test, y_lr_pred_test)
y_lr_test_rmse = root_mean_squared_error(y_test, y_lr_pred_test)

print("Mse on test data :", y_lr_test_mse)
print("R2 score on test data: ", y_lr_test_r2)
print("Rmse score on test data: ", y_lr_test_rmse)
```

```python
# to visualize in a dataframe. The transpose puts them in columns instead of rows.
lr_results = pd.DataFrame(["Linear Regression", y_lr_train_mse, y_lr_train_rmse, y_lr_train_r2, y_lr_test_mse, y_lr_test_rmse, y_lr_test_r2 ]).transpose()
lr_results.columns = (["Method", "Train(mse)", "Train(rmse)", "Train(r2)", "Test(mse)", "Test(rmse)", "Test(r2)"])
```
